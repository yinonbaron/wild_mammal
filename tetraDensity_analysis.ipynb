```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```python
data = pd.read_excel('tetraDensity.xls')
mammals = data[data['Class'] == 'Mammalia']
mammals['Binomial'] = mammals['Genus'].str.cat(mammals['Species'], sep=" ")
plt.figure()
ax = plt.gca()
#mammals.groupby('Binomial')['Density'].count().plot.hist(ax=ax, bins=100)
np.log10(mammals['Density']).plot.hist(ax=ax, bins=20)
#ax.set_xscale('log')
```

```python
mass_data = pd.read_csv('PanTHERIA_1-0_WR05_Aug2008.csv',sep='\t')
merged_data = mammals.merge(mass_data,left_on='Binomial',right_on='MSW05_Binomial')
plt.figure()
ax= plt.gca()
plt.scatter(np.log10(merged_data['5-1_AdultBodyMass_g']),np.log10(merged_data['Sampling_Area']))
```

```python
merged_data['Binomial'].unique().shape
#mammals['Binomial'].unique().shape
#mammals[['Genus','Species']].drop_duplicates()
merged_data.loc[np.argmax(merged_data['Density'])]

fit_data = np.log10(merged_data[['5-1_AdultBodyMass_g','Sampling_Area','Density']]).dropna()
#X = fit_data[['5-1_AdultBodyMass_g','Sampling_Area']]
#X = fit_data['5-1_AdultBodyMass_g']
X = fit_data['Sampling_Area']
X = sm.add_constant(X)
y = fit_data['Density']
x= sm.OLS(y,X).fit()
x.summary()
#plt.scatter(X,y)
```
