```python
import pandas as pd
import numpy as np
from sklearn import linear_model, metrics
```

```python
data = pd.read_csv('geb12617-sup-0001-suppinfo1 (1).csv')
mammals = data[(data['Taxonomic group'] =='Mammals') & (data['Diet (Proxy for Trophic level)'] =='Carnivore')]
mammals
```

```python
lm = linear_model.LinearRegression()
x_train = mammals[['Log10 body mass','Log10 study area']]
y_train = mammals['Log10 population density'] + mammals['Log10 study area']
lm.fit(x_train,y_train)
print(r'z = {} {}*x {}*y'.format(lm.intercept_,lm.coef_[0],lm.coef_[1]))
print('R^2 = {}'.format(metrics.r2_score(y_train,lm.predict(x_train))))
```
