```python
import pandas as pd
import numpy as np
```

```python
range_data = pd.read_csv('range_data.csv')
mass_data = pd.read_csv('PanTHERIA_1-0_WR05_Aug2008.csv', sep='\t')
range_data = range_data
range_data
```

```python
merged_df = mass_data.merge(range_data, how='inner', left_on='MSW05_Binomial', right_on='binomial') 
merged_df
```

```python
#merged_data[merged_data['binomial'] == 'Camelus ferus']
shai_func = lambda x: (10**(-0.822*np.log10(x) + 1.936))/1e4
shai_real_func = lambda x,y:  10**(1.65-0.53*np.log10(x) + 0.73*np.log10(y/1e4))
```

```python
merged_df['Log log total num'] = shai_func(merged_df['5-1_AdultBodyMass_g'])*merged_df['Range_size']
merged_df['Shai total num'] = shai_real_func(merged_df['5-1_AdultBodyMass_g'],merged_df['Range_size'])
```

```python
merged_df['Log log total mass'] = merged_df['Log log total num']*merged_df['5-1_AdultBodyMass_g']
merged_df['Shai total mass'] = merged_df['Shai total num']*merged_df['5-1_AdultBodyMass_g']
```

```python
top_taxa = merged_df[merged_df['Shai total mass'] > np.percentile(merged_df['Shai total mass'].dropna(),98)]
```

```python
top_taxa.to_excel('Top taxa.xlsx')
```
